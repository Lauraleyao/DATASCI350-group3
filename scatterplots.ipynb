```python
import wbgapi as wb
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('data/qtm350_final.csv')
print(data.head())
data
```

```python
#averages table
avg = data[data['Region'].isin(['Sub-Saharan Africa', 'South Asia'])]
region_year_avgs = avg.groupby(['Region', 'Year']).mean(numeric_only=True).reset_index()
region_year_avgs
```

```python
#Life Expectancy Over Time Scatterplot

sns.set(style="whitegrid")
plt.figure(figsize=(12, 6))

sns.scatterplot(
    data=region_year_avgs,
    x='Year',
    y='life_expectancy_birth',
    hue='Region',
    s = 200,
    alpha=0.7,
)

sns.lineplot(
    data=region_year_avgs,
    x='Year',
    y='life_expectancy_birth',
    hue='Region',
    estimator=None,
    lw=2,
    legend=False
)

plt.xticks(
    ticks=sorted(region_year_avgs['Year'].unique()),
    labels=sorted(region_year_avgs['Year'].unique()),
    rotation=45
)

plt.title("Life Expectancy Over Time in Sub-Saharan Africa")
plt.xlabel("Year")
plt.ylabel("Life Expectancy at Birth (Years)")
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()
plt.show()

#Primary School Enrollment Over Time Scatterplot

sns.set(style="whitegrid")
plt.figure(figsize=(12, 6))

sns.scatterplot(
    data=region_year_avgs,
    x='Year',
    y='primary_school_enrollment',
    hue='Region',
    s = 200,
    alpha=0.7,
)

sns.lineplot(
    data=region_year_avgs,
    x='Year',
    y='primary_school_enrollment',
    hue='Region',
    estimator=None,
    lw=2,
    legend=False
)

plt.xticks(
    ticks=sorted(region_year_avgs['Year'].unique()),
    labels=sorted(region_year_avgs['Year'].unique()),
    rotation=45
)

plt.title("Primary School Enrollment Over Time in Sub-Saharan Africa")
plt.xlabel("Year")
plt.ylabel("Primary School Enrollment")
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()
plt.show()

#GDP Over Time Scatterplot

sns.set(style="whitegrid")
plt.figure(figsize=(12, 6))

sns.scatterplot(
    data=region_year_avgs,
    x='Year',
    y='gdp_per_capita_constant_usd',
    hue='Region',
    s = 200,
    alpha=0.7,
)

sns.lineplot(
    data=region_year_avgs,
    x='Year',
    y='gdp_per_capita_constant_usd',
    hue='Region',
    estimator=None,
    lw=2,
    legend=False
)

plt.xticks(
    ticks=sorted(region_year_avgs['Year'].unique()),
    labels=sorted(region_year_avgs['Year'].unique()),
    rotation=45
)

plt.title("GDP Per Capita Over Time in Sub-Saharan Africa")
plt.xlabel("Year")
plt.ylabel("GDP Per Capita (USD)")
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()
plt.show()

#Basic Water Access Over Time Scatterplot

sns.set(style="whitegrid")
plt.figure(figsize=(12, 6))

sns.scatterplot(
    data=region_year_avgs,
    x='Year',
    y='basic_water_access_pct',
    hue='Region',
    s = 200,
    alpha=0.7,
)
sns.lineplot(
    data=region_year_avgs,
    x='Year',
    y='basic_water_access_pct',
    hue='Region',
    estimator=None,
    lw=2,
    legend=False
)
plt.xticks(
    ticks=sorted(region_year_avgs['Year'].unique()),
    labels=sorted(region_year_avgs['Year'].unique()),
    rotation=45
)

plt.title("Basic Water Access Over Time in Sub-Saharan Africa")
plt.xlabel("Year")
plt.ylabel("Basic Water Access (%)")
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()
plt.show()

#Basic Sanitation Access Over Time Scatterplot

sns.set(style="whitegrid")
plt.figure(figsize=(12, 6))

sns.scatterplot(
    data=region_year_avgs,
    x='Year',
    y='basic_sanitation_access_pct',
    hue='Region',
    s = 200,
    alpha=0.7,
)
sns.lineplot(
    data=region_year_avgs,
    x='Year',
    y='basic_sanitation_access_pct',
    hue='Region',
    estimator=None,
    lw=2,
    legend=False
)
plt.xticks(
    ticks=sorted(region_year_avgs['Year'].unique()),
    labels=sorted(region_year_avgs['Year'].unique()),
    rotation=45
)

plt.title("Basic Sanitation Access Over Time in Sub-Saharan Africa")
plt.xlabel("Year")
plt.ylabel("Basic Sanitation Access (%)")
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()
plt.show()

```

```python
#Top 3 and Bottom 3 Countries by GDP per Capita in Sub-Saharan Africa
ssa = data[data['Region'].isin(['Sub-Saharan Africa'])]

latest = (
    ssa.sort_values('Year')
      .groupby(['Country', 'Region'])
      .tail(1)
)
top3_ssa = (
    latest.sort_values(by=['Region', 'gdp_per_capita_constant_usd'], ascending=[True, False])
          .groupby('Region')
          .head(3)
)

bottom3_ssa = (
    latest.sort_values(by=['Region', 'gdp_per_capita_constant_usd'], ascending=[True, True])
          .groupby('Region')
          .head(3)
)

print("Top 3 Countries by GDP per Region:")
print(top3_ssa[['Region','Country','gdp_per_capita_constant_usd']])

print("\nBottom 3 Countries by GDP per Region:")
print(bottom3_ssa[['Region','Country','gdp_per_capita_constant_usd']])
```

```python
ssa_countries = pd.concat([top3_ssa, bottom3_ssa])['Country'].unique()
ssa_countries

#GDP per Capita
sns.set(style="whitegrid")
plt.figure(figsize=(12, 6))

sns.scatterplot(
    data=data[data['Country'].isin(ssa_countries)],
    x='Year',
    y='gdp_per_capita_constant_usd',
    hue='Country',
    s=200,
    alpha=0.7
)

sns.lineplot(
    data=data[data['Country'].isin(ssa_countries)],
    x='Year',
    y='gdp_per_capita_constant_usd',
    hue='Country',
    estimator=None,
    lw=2,
    alpha=0.8,
    legend=False
)

plt.xticks(sorted(data['Year'].unique()), rotation=45)
plt.title("GDP per Capita (Constant USD) Over Time for Top 3 and Bottom 3 GDP Countries in Sub-Saharan Africa")
plt.xlabel("Year")
plt.ylabel("GDP per Capita (Constant USD)")
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()
plt.show()

#Life Expectancy
sns.set(style="whitegrid")
plt.figure(figsize=(12, 6))

sns.scatterplot(
    data=data[data['Country'].isin(ssa_countries)],
    x='Year',
    y='life_expectancy_birth',
    hue='Country',
    s=200,
    alpha=0.7
)

sns.lineplot(
    data=data[data['Country'].isin(ssa_countries)],
    x='Year',
    y='life_expectancy_birth',
    hue='Country',
    estimator=None,
    lw=2,
    alpha=0.8,
    legend=False
)

plt.xticks(sorted(data['Year'].unique()), rotation=45)
plt.title("Life Expectancy Over Time for Top 3 and Bottom 3 GDP Countries in Sub-Saharan Africa")
plt.xlabel("Year")
plt.ylabel("Life Expectancy at Birth (Years)")
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()
plt.show()

#Primary School Enrollment
sns.set(style="whitegrid")
plt.figure(figsize=(12, 6))

sns.scatterplot(
    data=data[data['Country'].isin(ssa_countries)],
    x='Year',
    y='primary_school_enrollment',
    hue='Country',
    s=200,
    alpha=0.7
)

sns.lineplot(
    data=data[data['Country'].isin(ssa_countries)],
    x='Year',
    y='primary_school_enrollment',
    hue='Country',
    estimator=None,
    lw=2,
    alpha=0.8,
    legend=False
)

plt.xticks(sorted(data['Year'].unique()), rotation=45)
plt.title("Primary School Enrollment Over Time for Top 3 and Bottom 3 GDP Countries in Sub-Saharan Africa")
plt.xlabel("Year")
plt.ylabel("Primary School Enrollment (%)")
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()
plt.show()

#Basic Water Access
sns.set(style="whitegrid")
plt.figure(figsize=(12, 6))

sns.scatterplot(
    data=data[data['Country'].isin(ssa_countries)],
    x='Year',
    y='basic_water_access_pct',
    hue='Country',
    s=200,
    alpha=0.7
)

sns.lineplot(
    data=data[data['Country'].isin(ssa_countries)],
    x='Year',
    y='basic_water_access_pct',
    hue='Country',
    estimator=None,
    lw=2,
    alpha=0.8,
    legend=False
)

plt.xticks(sorted(data['Year'].unique()), rotation=45)
plt.title("Basic Water Access Over Time for Top 3 and Bottom 3 GDP Countries in Sub-Saharan Africa")
plt.xlabel("Year")
plt.ylabel("Basic Water Access (%)")
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()
plt.show()

#Basic Sanitation Access
sns.set(style="whitegrid")
plt.figure(figsize=(12, 6))

sns.scatterplot(
    data=data[data['Country'].isin(ssa_countries)],
    x='Year',
    y='basic_sanitation_access_pct',
    hue='Country',
    s=200,
    alpha=0.7
)

sns.lineplot(
    data=data[data['Country'].isin(ssa_countries)],
    x='Year',
    y='basic_sanitation_access_pct',
    hue='Country',
    estimator=None,
    lw=2,
    alpha=0.8,
    legend=False
)

plt.xticks(sorted(data['Year'].unique()), rotation=45)
plt.title("Basic Sanitation Access Over Time for Top 3 and Bottom 3 GDP Countries in Sub-Saharan Africa")
plt.xlabel("Year")
plt.ylabel("Basic Sanitation Access (%)")
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()
plt.show()

```
