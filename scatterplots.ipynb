```python
import wbgapi as wb
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('data/qtm350_final.csv')

print(data.head())
data
```

```python
#averages table
avg = data[data['Region'].isin(['Sub-Saharan Africa', 'South Asia'])]
region_year_avgs = avg.groupby(['Region', 'Year']).mean(numeric_only=True).reset_index()
region_year_avgs
```

```python
#GDP against LE

sns.set(style="whitegrid")
plt.figure(figsize=(12, 6))

sns.scatterplot(
    data=region_year_avgs,
    x='Year',
    y='life_expectancy_birth',
    hue='Region',
    size='gdp_per_capita_constant_usd',
    sizes=(50, 400),
    alpha=0.7,
)

plt.xticks(
    ticks=sorted(region_year_avgs['Year'].unique()),
    labels=sorted(region_year_avgs['Year'].unique()),
    rotation=45
)

plt.title("Life Expectancy Over Time, plotted against GDP per Capita")
plt.xlabel("Year")
plt.ylabel("Life Expectancy at Birth (Years)")
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()
plt.show()

```

```python
#water access against LE

sns.set(style="whitegrid")
plt.figure(figsize=(12, 6))

sns.scatterplot(
    data=region_year_avgs,
    x='Year',
    y='life_expectancy_birth',
    hue='Region',
    size='basic_water_access_pct',
    sizes=(10, 400),
    alpha=0.7,
)

plt.xticks(
    ticks=sorted(region_year_avgs['Year'].unique()),
    labels=sorted(region_year_avgs['Year'].unique()),
    rotation=45
)

plt.title("Life Expectancy Over Time, plotted against basic_water_access_pct")
plt.xlabel("Year")
plt.ylabel("Life Expectancy at Birth (Years)")
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()
plt.show()

```

```python

ssa = data[data['Region'].isin(['Sub-Saharan Africa'])]

latest = (
    ssa.sort_values('Year')
      .groupby(['Country', 'Region'])
      .tail(1)
)

top3 = (
    latest.sort_values(by=['Region', 'gdp_per_capita_constant_usd'], ascending=[True, False])
          .groupby('Region')
          .head(3)
)


bottom3 = (
    latest.sort_values(by=['Region', 'gdp_per_capita_constant_usd'], ascending=[True, True])
          .groupby('Region')
          .head(3)
)

print("Top 3 Countries by GDP per Region:")
print(top3[['Region','Country','gdp_per_capita_constant_usd']])

print("\nBottom 3 Countries by GDP per Region:")
print(bottom3[['Region','Country','gdp_per_capita_constant_usd']])

```
